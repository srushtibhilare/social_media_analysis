```python
df = df.drop_duplicates(subset='post_id')
```

```python
text_cols = ['platform', 'content_type', 'topic', 'language', 'region']

for col in text_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
df['post_datetime'] = pd.to_datetime(df['post_datetime'], errors='coerce', dayfirst=True)

df = df.dropna(subset=['post_datetime'])
```

```python
df["post_year"] = df["post_datetime"].dt.year 
df["post_month"] = df["post_datetime"].dt.month
df["post_day"] = df["post_datetime"].dt.day
df['post_hour'] = df['post_datetime'].dt.hour
```

```python
df["hashtag"] = df["hashtags"].fillna('').str.lower()

df['hashtag_count'] = df['hashtags'].apply(lambda x: len(x.split()) if x != '' else 0)
```

```python
num_cols = ['views', 'likes', 'comments', 'shares']

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(0)
    df[col] = df[col].clip(lower=0)
```

```python
df['engagement_rate'] = (
    df['likes'] + df['comments'] + df['shares']
) / df['views'].replace(0, np.nan)

df['engagement_rate'] = df['engagement_rate'].fillna(0)
```

```python
df['sentiment_score'] = pd.to_numeric(df['sentiment_score'], errors='coerce')
df['sentiment_score'] = df['sentiment_score'].clip(-1, 1)
df['sentiment_score'] = df['sentiment_score'].fillna(0)
```

```python
df = df.dropna(subset=['is_viral'])
df['is_viral'] = df['is_viral'].astype(int)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['platform'] = df['platform'].replace({
    'youtube shorts': 'youtube_shorts'
})
```

```python
df['region'] = df['region'].replace({
    'uk': 'united_kingdom',
    'us': 'united_states'
})
```

```python
df = df.drop(columns=['hashtag'])
```

```python
df['hashtags'] = df['hashtags'].apply(
    lambda x: [tag.replace('#', '') for tag in x.split()]
)
```

```python
df['engagement_rate'] = df['engagement_rate'].clip(0, 1)
```

```python
df.info()
```

```python
int_cols = ['views', 'likes', 'comments', 'shares', 'is_viral']
df[int_cols] = df[int_cols].astype(int)
```

```python
df.to_csv("cleaned_social_media_data.csv", index=False)
```

```python
df.head()
```

```python
df['total_engagement'] = df['likes'] + df['comments'] + df['shares']
```

```python
df['engagement_per_1k_views'] = (df['total_engagement'] / df['views']) * 1000
```

```python
df['sentiment_label'] = df['sentiment_score'].apply(
    lambda x: 'positive' if x > 0.2 else 'negative' if x < -0.2 else 'neutral'
)
```

```python
df_ml = pd.get_dummies(
    df,
    columns=['platform', 'content_type', 'topic', 'language', 'region'],
    drop_first=True
)
```

```python
df.to_csv("cleaned_social_media_posts.csv", index=False)
```

```python
import pandas as pd
import psycopg2

df = pd.read_csv(r"C:\Users\srush\OneDrive\Documents\final_social_media_analytics_dataset.csv")

conn = psycopg2.connect(
    host="localhost",
    database="raw",
    user="postgres",
    password="Camlin@02",
    port=5432
)
cur = conn.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS social_media_posts (
    post_id VARCHAR(50) PRIMARY KEY,
    platform VARCHAR(50),
    content_type VARCHAR(50),
    topic VARCHAR(50),
    language VARCHAR(10),
    region VARCHAR(100),
    post_datetime DATE,
    hashtags TEXT,
    views BIGINT,
    likes BIGINT,
    comments BIGINT,
    shares BIGINT,
    engagement_rate FLOAT,
    sentiment_score FLOAT,
    is_viral INT,
    post_year INT,
    post_month INT,
    post_day INT,
    post_hour INT,
    hashtag_count INT,
    total_engagement BIGINT,
    engagement_per_1k_views FLOAT,
    sentiment_label VARCHAR(20)
);
""")
conn.commit()

for i, row in df.iterrows():
    cur.execute("""
    INSERT INTO social_media_posts VALUES (
        %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
    )
    ON CONFLICT (post_id) DO NOTHING;
    """, tuple(row))

conn.commit()
cur.close()
conn.close()

print("Data uploaded successfully!")
```
